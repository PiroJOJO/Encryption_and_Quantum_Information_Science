# BB84

```python
import pandas as pd
db = pd.read_csv('export_bb84.csv')
db
```

```python
right_answer = len(db.drop(db[db['Sent'] != db['Received']].index, axis=0)) / len(db)
print(f'Количество совпадений {right_answer}')
```

# BB84 с фазовым кодированием

```python
import pandas as pd
db = pd.read_csv('export_bb84_f.csv')
db
```

```python
db['Sent'] = (((db['Alice'] == 0) & (db['Fa'].isin([0, 1]))).astype(int))
db
```

```python
right_answer = len(db.drop(db[db['Sent'] != db['result']].index, axis=0)) / len(db)
print(f'Количество совпадений {right_answer}')
```
